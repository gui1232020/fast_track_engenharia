```python
import pandas as pd
import requests
```

```python
def get_feriados_api(ano):
    url = f"https://brasilapi.com.br/api/feriados/v1/{ano}"
    
    response = requests.get(url)
    
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Erro ao buscar feriados de {ano}: {response.status_code}")
        return []

dados_2025 = get_feriados_api(2025)
dados_2026 = get_feriados_api(2026)

df_feriados = pd.DataFrame(dados_2025 + dados_2026)

print(f"Total de feriados: {len(df_feriados)}")
# print(df_feriados[['date', 'name']].head())
print(df_feriados[['date', 'name']])
```
