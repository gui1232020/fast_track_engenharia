### Import lib and define file source

```python
import pandas as pd
import pyarrow

file = '../../data/raw/jira_issues_raw.json'
```

### Read json file

```python
read_file = pd.read_json(file, typ='series')
read_file
```

### Get list of key issues

```python
df_bronze = pd.json_normalize(read_file.loc['issues'])
#df_bronze.head()
df_bronze
```

### Save bronze table in the parquet type

```python
df_bronze.to_parquet("../../data/bronze_layer/ingest_bronze.parquet", index=False)
```
